**The Lotka-Volterra equations**

The Lotka-Volterra equations are a mathematical model of predator--prey interactions between biological
species.  Let two variables `x` and `y` be proportional to the size of the populations of two species, traditionally called ``rabbits`` (the prey) and ``foxes`` (the predators).  You could think of `x` and `y` as being the population in thousands, say, so that `x=2` means there are 2000 rabbits.  Strictly the only allowed values of `x` and `y` would then be multiples of 0.001, since you can only have whole numbers of rabbits or foxes.  But 0.001 is a pretty close spacing of values, so it's a decent approximation to treat `x` and `y` as continuous real numbers so long as neither gets very close to zero.

In the Lotka--Volterra model the rabbits reproduce at a rate proportional to their population, but are eaten by the foxes at a rate proportional to both their own population and the population of foxes:

$\frac{dx}{dt} = \alpha x - \beta xy,$

where $\alpha$ and $\beta$ are constants.  At the same time the foxes reproduce at a rate proportional the rate at which they eat rabbits---because they need food to grow and reproduce---but also die of old age at a rate proportional to their own population:

$\frac{dy}{dt} = \gamma xy - \delta y,$

where $\gamma$ and $\delta$ are also constants.

  the same axes from $t=0$ to $t=30$.  (Hint: Notice that the differential

  notation, the right-hand side of each equation is a function $f(\vec{r})$

  and foxes.
\end{enumerate}

### Setting the constants and defining the functions

````julia
ùõº = 1
ùõΩ = 0.5
ùõæ = 0.5
ùõø = 2

function f(r,t)
    x = r[1]
    y = r[2]
    fx = ùõº*x - ùõΩ*x*y
    fy = ùõæ*x*y - ùõø*y
return Array([fx,fy])
end
````

### Setting up the Rungge-Kutta 4th Order method to solve the ODE

````julia
a = 0.0
b = 30.0
N = 1000
h = (b-a)/N
r = Array([2.0,2.0])
tpoints = range(a,stop=b,step=h)

function lotkavolterra(tpoints)
    r = Array([2.0,2.0]) #initial condition
    xpoints = []
    ypoints = []
    for t in tpoints
        append!(xpoints, r[1])
        append!(ypoints, r[2])
        k1 = h*f(r,t)
        k2 = h*f(r+0.5*k1,t+0.5*h)
        k3 = h*f(r+0.5*k2,t+0.5*h)
        k4 = h*f(r+k3,t+h)
        r += (k1+2*k2+2*k3+k4)/6
    end
    return xpoints, ypoints
end
````

````julia
rabbits, foxes = lotkavolterra(tpoints);
time_1 = @elapsed lotkavolterra(tpoints)
````

````julia
plot(tpoints, rabbits, label="rabbits", lc = "magenta");
plot!(tpoints, foxes, label="foxes", lc = "green", xlabel = "t", ylabel = "population in 1000")
````

The Lotka-Volterra Model describes the simplest case of a predator-prey relationship. The cyclical behaviour of rabbits and foxes is evidents, and the fox population increase tracked the peaks of the rabbit population. This means that further increase in the number of foxes will compromise the amount of prey, maybe due to competition. At some point, the continuous decline of the number of rabbits will cause a decline in the number of foxes. Then, the continuous decline on the number of foxes, will at some point, cause an increase in the rabbit population, and so on and so forth. However, this model is too simplistic since it doesn't take into account the complexity of a realistic situation, such as the randomness of the predator-prey encounters through a homogeneous environment.

Source: M. Beals, L. Gross, and S. Harrell, PREDATOR-PREY DYNAMICS: LOTKA-VOLTERRA, The Institute of Environmental Modeling, The University of Tennessee, Knoxville (1999).
URL: http://www.tiem.utk.edu/~gross/bioed/bealsmodules/predator-prey.html (Date retrieved: May 5, 2018)


````julia
plot(rabbits,foxes, xlabel="rabbits", ylabel="foxes")
````

According to Sharov (1996), This model has no asymptotic stability, it does not converge to an attractor (does not "forget" initial conditions).

"This figure shows relative changes in prey predator density for both initial conditions. Trajectories are closed lines.
The model of Lotka and Volterra is not very realistic. It does not consider any competition among prey or predators. As a result, prey population may grow infinitely without any resource limits. Predators have no saturation: their consumption rate is unlimited. The rate of prey consumption is proportional to prey density. Thus, it is not surprising that model behavior is unnatural showing no asymptotic stability. However numerous modifications of this model exist which make it more realistic."

Sharov, A. Lotka-Volterra Model. Department of Mathematics, University of Texas at Austin (1996).

Here we use a documentation of Lotka-Voltera equations which uses the `DifferentialEquations` package and we compare it with our custom Runge-Kutta solver.

````julia
function lotka_volterra(du,u,p,t)
  x, y = u
  Œ±, Œ≤, Œ¥, Œ≥ = p
  du[1] = dx = Œ±*x - Œ≤*x*y
  du[2] = dy = -Œ¥*y + Œ≥*x*y
end
u0 = [2.0,2.0]
tspan = (0.0,30.0)
p = [1.0,0.5,2.0,0.5]
prob = ODEProblem(lotka_volterra,u0,tspan,p,NSTEPS=1000)
````

````julia
sol = solve(prob)
plot(sol, lw = 3)
````

Overlaying the results of the rabbit and fox population obtained using Runge-Kutta and the solution to the ODE using the `ODEPoblem` solver, we can confidently say  that the results yielded were accurate.

````julia
plot!(tpoints, rabbits, lw = 2, label="rabbits", lc = "magenta", s = :dash);
plot!(tpoints, foxes, lw = 2, label="foxes", lc = "green", s = :dash, ylabel = "population in 1000")
````

````julia
time_2 = @elapsed ODEProblem(lotka_volterra,u0,tspan,p,NSTEPS=1000)
````

In terms of runtime,

````julia
print("The time elapsed to solve the ODE using the DifferentialEquations package is $(time_1/time_2)x")
print(" faster than the Runge-Kutta method.")
````
