## KR2

\textbf{Demonstrate that Julia is able to determine constants in codes.}

```julia
@code_warntype 14^pi
```

```julia
f(x) = log(10.0)*x
```

```julia
@code_llvm f(4.0)
```

This `   %1 = fmul double %0, 0x40026BB1BBB55516` represents the operation and `0x40026BB1BBB55516` is the code of the constant `log(10.0)`. Instead of performing the log operation, it just finds the value from it's database. In short, Julia is able to know where are the constants in our code and apply it to reduce the number of operations.

```julia
g(x) = x*log(x)
```

```julia
@code_llvm g(4.0)
```

In this particular function where constant is non-existent, Julia is forced to call the log operation as indicated by `  %1 = call double @j_log_2551(double %0) #0` before it executes the multiplication operation.

\textbf{Demonstrate Julia's type-inference and multiple dispatch.}


In this portion, we explore functions and methods, on how to create them and how they are used. Before creating one, it's better to check if the function already exists so that we can avoid overloading existing functions. For example, is there an existing function called `permute`?

```julia
?permute
```

If there isn't or the function you queried is slightly incorrect, Julia's help mode will give you suggestions. On my query `?permute`, my syntax is incorrect hence it suggested the following alternatives: `permute! permutedims permutedims! PermutedDimsArray invpermute!`.

```julia
?permute!;
```

```julia
permute!([1, 1, 3, 4],[2, 4, 3, 1]) 
```

If a function that we deem doesn't exist, then we can define a general one and specialized one. From the discussion, these were the examples on how to set up a custom function.

```julia
what(x) = "some type $(typeof(x))"
what(x::String) = "a string"
what(x::Number) = "a number"
what(x::Real) = "a real number"
what(x::Rational) = "a rational number"
what(x::Int) = "an integer"
what(x::Complex) = "a complex number"
```

```julia
?what
```

It essentially demonstrates how Julia is aware of the type of a given input, hence "type-inference". In the function crafted, different inputs correspond to difference outputs but it is contained in a singular function. When called, it decipheres the type of the input as shown by some demonstrations below.

```julia
println("4.0 is $(what(4.0)).")
println("4+2im is $(what(4+2im)).")
println("4 is $(what(4)).")
```

## KR4

\textbf{ Show the difference, if any, between your own sum function my_sum(x::Vector) and @time. Use a for-loop for your customized sum function.}

Here's a customized sum function `my_sum(::Vector)` demonstrated during class.

```julia
function my_sum(x::Vector)
    s = zero(eltype(x))
    for el in x
        s += el
    end
    return s
end
```

The main goal for KR4 is to show how a custom function compares with the defualt sum function of julia by checking the runtime (`@time`).

```julia
x = rand(10_000_000);
```

```julia
@time my_sum(x)
```

```julia
@time my_sum(x)
```

```julia
?sum(::AbstractArray)
```

```julia
@time sum(x)
```

```julia
@time sum(x)
```

## KR7

```julia
iter = 200 
time = zeros(iter)
n_list = range(10, length = iter, step=10)
for i in range(1, stop = iter)
    tt = @timed plotMandelbrot( (-1.0,1.0); n=n_list[i] )
    time[i] = tt[2]
end
```

```julia
plot(n_list, time)#, seriestype = :scatter, title = "Julia's computational complexity of the Mandelbrot")
```
