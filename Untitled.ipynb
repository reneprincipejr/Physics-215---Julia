```julia
using Profile
using ProfileView
using Plots
```

```julia
using DifferentialEquations
```

```julia
function lotka_volterra(du,u,p,t)
  x, y = u
  α, β, δ, γ = p
  du[1] = dx = α*x - β*x*y
  du[2] = dy = -δ*y + γ*x*y
end
u0 = [2.0,2.0]
tspan = (0.0,30.0)
p = [1.0,0.5,2.0,0.5]
prob = ODEProblem(lotka_volterra,u0,tspan,p,NSTEPS=1000)
```

```julia
sol = solve(prob)
plot(sol, lw = 3)
```

```julia
@profile ODEProblem(lotka_volterra,u0,tspan,p,NSTEPS=1000);
```
